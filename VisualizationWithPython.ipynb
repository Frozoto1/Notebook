```python
fig = plt.figure(1)
box = dict(facecolor='yellow', alpha=0.2)
ax1 = fig.add_subplot(211)
ax1.text(0.2,0.8,'hello')
#
ax1.yaxis.tick_right()
ax1.yaxis.set_label_position("right")
ax1.set_ylabel('right\nVer',fontsize=20,bbox=box,rotation=-90,labelpad=50)
#
ax2 = fig.add_subplot(212)
ax2.text(0.2,0.8,'hello')
brect = plt.Rectangle((0.2,0.25),0.2,0.5,color='red',fill=False)
ax2.add_patch(brect)
ylabel2 = ax2.set_ylabel('left\nHor',fontsize=20,rotation=0,labelpad=70,bbox=box)
#
plt.axes([0.6, 0.12, 0.2, 0.76])
plt.xlim([0,1])
plt.ylim([0,1])
plt.show()
```

### 共享x,y轴

```python
fig, ax = plt.subplots(2,3,sharex=True,sharey=True)
for i in range(2):
    for j in range(3):
        ax[i,j].text(0.5,0.5,str((i,j)),fontsize=18,ha='center')
plt.show()
```

### 图形间距调整

```python
fig = plt.figure(figsize=[5,5])
#
fig.add_subplot(221)
plt.text(0.5,0.4,'bottom=0.2,top=0.4',ha='center',fontsize=10)
#
fig.add_subplot(222)
plt.text(0.5,0.4,'bottom=0.2,top=0.4',ha='center',fontsize=10)
plt.subplots_adjust(wspace=0.2,hspace=0.2)
#
fig.add_subplot(223)
plt.text(0.5,0.4,'bottom=0.2,top=0.4',ha='center',fontsize=10)
plt.subplots_adjust(wspace=0.2,hspace=0.2)
#
fig.add_subplot(224)
plt.text(0.5,0.4,'bottom=0.2,top=0.4',ha='center',fontsize=10)
plt.subplots_adjust(wspace=0.2,hspace=0.2,bottom=0.2,top=0.4)
#
plt.savefig('test6.pdf')
plt.show()
```

### 自定义图形布局

```python
#实现不规则的多行多列子图
import matplotlib.pyplot as plt
grid = plt.GridSpec(2, 3, wspace=0.5, hspace=0.5)
plt.subplot(grid[0,0])
plt.subplot(grid[0,1:3])
plt.subplot(grid[1,0:2])
plt.subplot(grid[1,2])
plt.text(0.5,0.25,'hellooocarain',ha='center')
plt.show()
```

```python
plt.figure(figsize=[18,6])
ax1 = plt.subplot2grid((4, 3), (0, 0), rowspan=4)
ax2 = plt.subplot2grid((4, 3), (0, 1))
ax3 = plt.subplot2grid((4, 3), (1, 1))
ax4 = plt.subplot2grid((4, 3), (2, 1))
ax5 = plt.subplot2grid((4, 3), (3, 1))
ax6 = plt.subplot2grid((4, 3), (0, 2),rowspan=4)
plt.show()
```

### 关于修改字体

一般字体的修改都可以通过fontsize属性设置，下面介绍全局字体的设置:

[stackoverflow 问答](https://stackoverflow.com/questions/3899980/how-to-change-the-font-size-on-a-matplotlib-plot)

#### Ex1设置全局的统一字体格式

```python
font = {'family' : 'Arial',
        'weight' : 'bold',
        'size'   : 22}
matplotlib.rc('font', **font)
fig,ax = plt.subplots(nrows=1,ncols=1,figsize=[8,8])
#针对某一特定的属性通过fontsize设置字体大小，优先于全局设置
ax.text(0.1,0.2,'Hello',fontsize=15)
#for tick in ax.yaxis.get_major_ticks():
#    tick.label1.set_fontweight('bold')
plt.show()
```

#### Ex2 另一种设置方式，附加将图形轴加粗

```python
fig,ax = plt.subplots(ncols=1,nrows=1,figsize=[8,8])
ax.plot([1,2,3],[4,5,2],label='A')
ax.plot([1,2,3],[3,6,1],label='B')
#
SMALL_SIZE = 8
MEDIUM_SIZE = 15
BIGGER_SIZE = 20
#
plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('axes', linewidth=6)
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
#
ax.text(1.5,3.5,'Hello Oocarain')
ax.set_title('Test title')
ax.set_xlabel('Test_X_label')
ax.set_ylabel('Test_Y_label')
ax.legend()
plt.show()
```
