```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from mpl_toolkits.basemap import Basemap
```

```python
 # Make a data frame with the GPS of a few cities:
plt.figure(figsize=[12,8])
data = pd.DataFrame({
'lat':[-34, 49, -38, 59.93, 5.33, 45.52, -1.29, -12.97, 39.9],
'lon':[-58, 2, 145, 30.32, -4.03, -73.57, 36.82, -38.5, 116.4],
'name':['Buenos Aires', 'Paris', 'melbourne', 'St Petersbourg', 'Abidjan', 'Montreal', 'Nairobi', 'Salvador', 'Beijing']
})
 
# A basic map
m=Basemap(llcrnrlon=-180,urcrnrlon=180, llcrnrlat=-90,urcrnrlat=90)
m.drawmapboundary(fill_color='#A6CAE0', linewidth=2)
m.fillcontinents(color='grey', alpha=0.8, lake_color='grey')
m.drawcoastlines(linewidth=0.1, color="white")
 
# Add a marker per city of the data frame!
m.plot(data['lon'], data['lat'], linestyle='none', marker="o", markersize=16, alpha=0.6, c="orange", markeredgecolor="black", markeredgewidth=1)
plt.show()
```
