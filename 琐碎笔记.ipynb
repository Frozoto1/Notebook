## R与Python共用

```python
%load_ext rpy2.ipython

%R require(ggplot2)

import pandas as pd
df = pd.DataFrame({ 'Letter': ['a', 'a', 'a', 'b', 'b', 'b', 'c', 'c', 'c'], 'X': [4, 3, 5, 2, 1, 7, 7, 5, 9], 'Y': [0, 4, 3, 6, 7, 10, 11, 9, 13], 'Z': [1, 2, 3, 1, 2, 3, 1, 2, 3] })
```

```python
df
```

```python
%%R -i df
ggplot(data = df) + geom_point(aes(x = X, y= Y, color = Letter, size = Z))
```

## 读取多个文件创建单个dataframe

```python
import glob
```

**glob模块是python中类似Unix风格的路径名模式扩展，可以很方便的获取文件名或路径名**

[Python glob模块 文档](https://docs.python.org/3/library/glob.html)

```python
files = glob.glob('file_*.csv')
result = pd.concat([pd.read_csv(f) for f in files],ignore_index=True)
result
```

```python
#删除第一列的内容
result.drop(['Unnamed: 0'],axis=1,inplace=True)
```
